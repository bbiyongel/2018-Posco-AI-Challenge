```python
sub_result = []
for sheet in xl.sheet_names[1:]:
    raw = xl.parse(sheet)
    raw = raw.drop([0,1,2],axis=0)
    for col_num in range(len(raw.columns) // 4 + 1):
        st = col_num*(col_len + 1)
        et = st + col_len
        print(st, et)
        sub = raw.iloc[:,st:et]
        sub.dropna(how='all',inplace=True)
        sub.drop([sub.index[-1]],axis=0,inplace=True)
        sub.iloc[:,0].fillna(method='ffill',inplace=True)
        sub.iloc[:,2].fillna(value='test',inplace=True)
        sub.iloc[:,1].fillna(value=-1,inplace=True)
        sub.columns = ['day','swell','time']
        sub.reset_index(drop=True,inplace=True)
        sub.info()

        sub_result.append(sub)
```

```python
merged_df = pd.concat(sub_result,axis=0).reset_index(drop=True)
merged_df.info()
merged_df.head()
```
