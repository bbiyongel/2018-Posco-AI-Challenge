```python
import pandas as pd
import glob
```

```python
sub_info = []
sub_info.append(glob.glob('data/MARINE_CWBUOY_22453*'))
sub_info.append(glob.glob('data/MARINE_CWBUOY_22490*'))

sub_info.append(glob.glob('data/20180712142659/MARINE_BUOY_21229*'))
sub_info.append(glob.glob('data/20180712142659/MARINE_BUOY_22105*'))
sub_info.append(glob.glob('data/20180712142659/MARINE_BUOY_22106*'))
sub_info.append(glob.glob('data/20180712142659/MARINE_BUOY_22190*'))
sub_info.append(glob.glob('data/MARINE_BUOY_22189*'))
```

```python
sub_df_list = []
keys = []
for info in sub_info:
    files = []
    pos = 0

    for filename in info:
        print(filename)
        sub_df = pd.read_csv(filename,encoding='euc-kr')
        files.append(sub_df)
        pos = str(sub_df['지점'][0])
        print(pos)

    sub_df = pd.concat(files,axis=0)
    sub_df.sort_values('일시',inplace=True)
    sub_df.reset_index(inplace=True, drop=True)
    sub_df.drop(['지점'],axis=1,inplace=True)
    temp_col_names = []

    temp_date = pos+'_'+'일시'
    keys.append(temp_date)
    
    for col in sub_df.columns:
        print(col)
        temp_col_names.append(pos+'_'+str(col))
            
    sub_df.columns = temp_col_names                        
    sub_df.info()
    
    sub_df[temp_date] = sub_df[temp_date].apply(lambda x:x.replace('-','').replace(' ','').replace(':','')[:-2])
    sub_df.drop_duplicates(subset=[temp_date],keep='first',inplace=True)
    
    sub_df_list.append(sub_df)
```

```python
df = pd.read_csv('./swell_data_by_hour.csv')
df['hour'] = df['hour'].astype(str)
df.info()
```

```python
df.head()
```

```python
merged_df = df.copy()
for i,sub_df in enumerate(sub_df_list):
    merged_df = merged_df.merge(sub_df, left_on='hour',right_on=keys[i],how='left')
    merged_df.info()
```
