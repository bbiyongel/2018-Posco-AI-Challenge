```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./merged_info_data.tsv',sep='\t',index_col='Unnamed: 0')
df.info()
```

```python
df.corr()
```

```python
df[df['swell'].isnull()]
```

```python
df['swell(t+1)'] = df['swell'].shift(-1)
df.info()
```

```python
df.corr()['swell(t+1)']
```

```python
df.columns
```

```python
features = ['weather_deterioration', 'swell','22453_최대파고(m)', '22453_유의파고(m)', '22453_평균파고(m)',
       '22453_파주기(sec)']
target = ['swell(t+1)']
```

```python
train = df[features + target].dropna()
train.info()
```

```python
X = train[features].values
y = train[target].values
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC
from sklearn.metrics import log_loss, accuracy_score
from sklearn.preprocessing import minmax_scale
```

```python
X = minmax_scale(X,feature_range=(0,1))
```

```python
svc = LinearSVC()
svc
```

```python
cross_val_score(svc,X=X,y=y.reshape(-1,),cv=5)
```

```python
df['swell'].value_counts()
```

```python
svc.fit(X,y)
```

```python
pred = svc.predict(X)
true = y
```

```python
log_loss(true,pred)
```

```python
accuracy_score(true,pred)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y)
```

```python
print(train_X.shape,test_X.shape, train_y.shape, test_y.shape)
```

```python
svc = LinearSVC()
svc.fit(train_X,train_y)
```

```python
pred = svc.predict(test_X)
true = test_y
```

```python
log_loss(true,pred)
```

```python
accuracy_score(true,pred)
```
