```python
def day2hour(day,swell,time):
    day_today = int(day.strftime('%Y%m%d%H'))
    day_tomorrow = int((day + timedelta(days=1)).strftime('%Y%m%d%H'))
    
    result = []
    
    if time == 'test':
        for hour in range(7,24):
                result.append((day_today + hour,'?'))
            
        for hour in range(0,7):
                result.append((day_tomorrow + hour,'?'))
    
        return result
    

    hour_list = time.split('~')
    st = int(hour_list[0][:2])
    et = int(hour_list[1][:2])
    
    
    if st < 7: # tomorrow
        
        for hour in range(7,24):
            result.append((day_today + hour,0))
            
        for hour in range(0,st):
            result.append((day_tomorrow + hour,0))
        
        for hour in range(st,et):
            result.append((day_tomorrow + hour,swell))
        
        for hour in range(et,7):
            result.append((day_tomorrow + hour,0))
        
    
    else: # today
        
        if et <= 7: # today ~ tommorrow
            
            for hour in range(7,st):
                result.append((day_today + hour,0))
            
            for hour in range(st,24):
                result.append((day_today + hour,swell))
            
            for hour in range(0,et):
                result.append((day_tomorrow + hour,swell))
                
            for hour in range(et,7):
                result.append((day_tomorrow + hour,0))
                
            
        else: # today
            for hour in range(7,st):
                result.append((day_today + hour,0))
            
            for hour in range(st,et):
                result.append((day_today + hour,swell))
                
            for hour in range(et,24):
                result.append((day_today + hour,0))
                
            for hour in range(0,7):
                result.append((day_tomorrow + hour,0))
            
    return result
```

###  climate 분할
* -1: 기상악화만
* 0이상 : 기상학화 + swell
*  0 : 없음
* ? : test

```python
def get_swell(x):
    #print(x)
    x = str(x)
    
    if x == '?':
        return None
    
    x = float(x)
    
    if x == 0:
        return 0
    
    if x < 0:
        return 0
    
    if x > 0:
        return 1
```

```python
hour_df['weather_deterioration'] = hour_df['climate'].apply(lambda x: 0 if str(x) == '0.0' or str(x) == '0' or str(x) == '?' else 1)
hour_df['swell'] = hour_df['climate'].apply(lambda x: get_swell(x))
```
