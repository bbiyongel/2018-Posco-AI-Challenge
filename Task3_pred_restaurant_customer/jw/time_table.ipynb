```python
d1 = date(2003, 3, 1)  # start date
d2 = date(2017, 12, 31)  # end date

delta = d2 - d1         # timedelta

time_table = []
for i in range(delta.days + 1):
    time_table.append(d1 + timedelta(i))
```

```python
len(time_table)
```

```python
date_df = pd.DataFrame(time_table, columns=['date'])
date_df['date'] = date_df['date'].apply(lambda x:x.strftime("%Y%m%d")).astype(int)
date_df.info()
```

```python
date_df.head()
```

```python
train_df = pd.read_csv('./train_data.tsv',sep='\t')
train_df.info()
train_df.head()
```

```python
date_merge_df = date_df.merge(train_df,left_on='date',right_on='일자',how='outer')
date_merge_df.info()
```
